# Imports

Import matplotlib and pyplot along with numpy.

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

Use numpy to generate some sample data.

```python
x = np.linspace(0,99, num=100, dtype=int)
print(x)
```

# Nicer Plots

Use various different techniques to make nice plots.

## Regular Scale on Both Axes

```python
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(1,1,1)
ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.spines['top'].set_color("#aaaaaa")
ax.spines['right'].set_color("#aaaaaa")
# ax.spines['right'].set_visible(False)
ax.axis("square")
ax.grid()
ax.plot(x, x, marker="o", linestyle="None", markersize=2)
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
ax = f.add_subplot(1,1,1)
ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.spines['top'].set_color("#aaaaaa")
ax.spines['right'].set_color("#aaaaaa")
# ax.spines['right'].set_visible(False)
# ax.axis("square")
ax.grid()
ax.plot(x, x, marker="o", linestyle="None", markersize=2)
```

## Tickmarks

```python
# set up the plot
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
ax = f.add_subplot(1,1,1)
ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.axis("square")

# configure spines
ax.spines['top'].set_color("#aaaaaa")
ax.spines['right'].set_color("#aaaaaa")
ax.spines['left'].set(linewidth=1.2)
ax.spines['bottom'].set(linewidth=1.2)

# setup tick locations
ax.xaxis.set_major_locator(MultipleLocator(20))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))


# ax.set_xticks(np.linspace(0, 100, num=5, dtype=int))
# ax.set_yticks(np.linspace(0, 100, num=5, dtype=int))
# ax.minorticks_on()

ax.tick_params(which='minor', width=1.2, length=4, color="#000000")
ax.grid(which="minor", linewidth=1, color='#dddddd', zorder=-10, linestyle="--")

ax.tick_params(which='major', width=1.2, length=10, color="#000000")
ax.grid(which="major", linewidth=1.0, color='#aaaaaa', zorder=-10)


ax.plot(x, x, marker="o", linestyle="None", markersize=2)
```
